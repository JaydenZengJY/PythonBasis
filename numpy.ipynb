# 一、ndarray
## 1，ndarray的特性
### 多维性

```
import numpy as np
```

```
arr = np.array(5)  #创建0维的ndarray数组
print(arr)
print('arr的维度', arr.ndim) #数组的维度number of dimensions
```

```
arr = np.array([1, 2, 3])  #创建1维的ndarray数组
print(arr)
print('arr的维度', arr.ndim) #数组的维度number of dimensions
```

```
arr = np.array([[1, 2, 3], [4, 5, 6]])  #创建2维的ndarray数组
print(arr)
print('arr的维度', arr.ndim) #数组的维度number of dimensions
```

# 同质性

```
arr = np.array([1, 'hello'])  #不同的数据类型会被强制转换为相同的数据类型
print(arr)
```

```
arr = np.array([1, 2.5])  #不同的数据类型会被强制转换为相同的数据类型
print(arr)
```

## 2.ndarray的属性

```
arr = np.array(1)
print(arr)
print('数组的形状', arr.shape)
print('数组的形状', arr.ndim)
print('元素的个数', arr.size)
print('元素的数据类型', arr.dtype)
print('元素的转置', arr.T)
```

```
arr = np.array([1, 2.5, 3])
print(arr)
print('数组的形状', arr.shape)
print('数组的形状', arr.ndim)
print('元素的个数', arr.size)
print('元素的数据类型', arr.dtype)
```

```
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
print('数组的形状', arr.shape)
print('数组的形状', arr.ndim)
print('元素的个数', arr.size)
print('元素的数据类型', arr.dtype)
print('元素的转置\n', arr.T)
```

## 3.ndarray的创建

```
# 基础的创建方法
list1 = [1, 2, 3]
arr = np.array(list1, dtype=np.float64)
print(arr.ndim) #属性
print(arr)
```

```
# copy
arr1 = np.copy(arr) # 元素跟原始的数组相同，但不是一个数组
print(arr1)
arr1[0] = 5
print(arr1)
print(arr)
```

```
# 预定义形状
# 全0 全1 未初始化 固定值
# 全0
arr = np.zeros((2,3), dtype=int)
print(arr)
print(arr.dtype)
```

```
arr = np.zeros((2,), dtype=int)
print(arr)
print(arr.dtype)
```

```
# 全1
arr = np.ones((5,8), dtype=int)
print(arr)
```

```
# 未初始化（数随机）
arr = np.empty((4,3))
print(arr)
```

```
arr = np.full((3,4), 2026)
print(arr)
```

```
arr2 = np.ones_like(arr)
print(arr2)
arr3 = np.zeros_like(arr)
print(arr3)
arr4 = np.full_like(arr, 2026)
print(arr4)
```

```
# 等差数列 1 2 3 4 5
arr = np.arange(1,10,1) #start/ end/ step
print(arr)
```

```
# 等差数列 1 2 3 4 5
arr = np.arange(2,11,2) #start/ end/ step
print(arr)
```

```
# 等间隔数列
app = np.linspace(1, 10, 5)
print(app)
```

```
# 等间隔数列
app = np.linspace(0, 100, 5, dtype=int)
print(app)
arr = np.arange(0, 101, 25)
print(arr)
```

```
# 对数间隔数列
arr = np.logspace(0, 4, 3, base=3, dtype=int)
print(arr)
```

```
# 特殊矩阵
# 单位矩阵
arr =np.eye(3, dtype=int)
print(arr)
```

```
# 对角矩阵
arr =np.diag([1, 2, 3])
print(arr)
```

```
# 随机数组的生成
# 生成0-1之间的随机浮点数（均匀分布）
arr = np.random.rand(2, 3)
print(arr)
```

```
# 生成指定范围区间的随机浮点数
arr = np.random.uniform(3, 6, (2, 3))
print(arr)
```

```
# 生成指定范围区间的随机整数
arr = np.random.randint(3, 30, (2, 3))
print(arr)
```

```
# 生成随机数列（正态分布）
arr = np.random.randn(2, 3)
print(arr)
```

```
# 随机种子
np.random.seed(20)
arr = np.random.randint(1, 10, (2, 5))
print(arr)
```

### 5.索引与切片

```
# 一维数组的索引与切片
arr = np.random.randint(1, 100, 20)
```

```
print(arr)
```

```
print(arr[0])
print(arr[:]) #获取全部数据
print(arr[2:7]) #start:end+1(左包右不包)
print(arr[slice(2,15,2)]) #start.end,step
print(arr[(arr>10) & (arr<40)]) #布尔索引
```

```
# 二维数组的索引与切片
arr = np.random.randint(1, 100, (4,8))
print(arr)
```

```
print(arr[1,1]) #索引
print(arr[0:3,1])
print(arr[2][arr[2]>50])
print(arr[:,3])
```

### 6.ndarray的运算

```
# 算术的运算
a = np.array([1, 2, 3])
b = np.array([2, 2, 3])
print(a + b)
```

```
c = [1, 3, 3]
d = [2, 3, 4]
for i in range(len(c)):
    d[i] = d[i] + c[i]
print(d)
```

```
# 二维计算
a = np.array([[1, 4, 5],[6, 5, 8],[1, 2, 3]])
b = np.array([[2, 2, 3],[4, 7, 9],[9, 5, 4]])
print(a * b)
```

```
# 数组与标量计算
a = np.array([[1, 4, 5],[6, 5, 8],[1, 2, 3]])
print(a + 3)
```

```
# 广播机制:1.获取形状2/是否可广播
a = np.array([1, 2, 3]) #1*3
b = np.array([[2], [2], [3]]) #3*1
print(a * b)
```

```
# 矩阵运算
a = np.array([[1, 4, 5],[6, 5, 8],[1, 2, 3]])
b = np.array([[2, 2, 3],[4, 7, 9],[9, 5, 4]])
print(a @ b)
```

# 2。统计函数
## 求和，平均值，中位数，标准差，方差， 查找最大值、最小值，分位数，累计和，累计差

```
arr = np.random.randint(1, 20, 8)
print(arr)
```

```
# 求和
print(np.sum(arr))
```

```
# 求平均值
print(np.average(arr))
print(np.mean(arr))
```

```
# 计算中位数
print(np.median(arr))
print(np.median([1, 2, 3, 4]))
```

```
# 计算标准差和方差
print(np.var([1, 2, 3]))#方差
print(np.std([1, 2, 3]))#标准差
```

```
# 计算最大值和最小值
print(arr)
print(np.max(arr), np.argmax(arr))
print(np.min(arr), np.argmin(arr))
```

```
# 分位数
np.random.seed(0)
arr = np.random.randint(0, 100, 4)
print(arr)
print(np.median(arr))
print(np.percentile(arr, 20))
```

```
# 累积的和，累积的积
arr = np.array([1, 2, 3])
print(np.cumsum(arr))
print(np.cumprod(arr))
```

# 比较函数
### 比较是否大于、小于、等于
### 逻辑与、或、非
### 检查数组中是否有一个True,是否所有都为True,自定义条件

```
# 是否大于
print(np.greater([3, 4, 5, 6, 7], 4))
# 是否小于
print(np.less([3, 4, 5, 6, 7], 4))
# 是否等于
print(np.equal([3, 4, 5, 6, 7], 4))
print(np.equal([3, 4, 5], [6, 7, 8]))# 同形状
```

```
print(np.logical_and([0, 0], [1, 1]))
print(np.logical_or([0, 0], [1, 1]))
print(np.logical_not([1, 0]))
```

```
# 检查元素是否至少有一个元素为True
print(np.any([1, 0]))
# 检查元素是否全部为True
print(np.all([1, 0]))
```

```
# 自定义条件
# print(np.where(条件,符合条件,不符合条件))
arr = np.array([1, 2, 3, 4, 5])
print(np.where(arr>3, arr, 0))
```

```
score = np.random.randint(50, 100, 20)
print(score)
print(np.where(score>=60, '及格', '不及格'))
```

```
print(np.where(
    score<60,'不及格',np.where(
        score<80, '良好', '优秀'
    )
))
```

```
# np.select(条件， 返回的结果）
print(np.select([score>80,(score>60)&(score<=80),score<60],
                ['优秀','良好','不及格'],
                default='No'))
```

```
# 排序函数
np.random.seed(0)
arr = np.random.randint(1, 100, 20)
print(arr)
print(np.sort(arr))
print(np.argsort(arr))
print(arr)#不被污染
```

```
# 去重函数
print(np.unique(arr))
```

```
# 数组的拼接
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print(np.concatenate((arr1,arr2)))
```

```
# 数组的切割
print(np.split(arr,5))
print(np.split(arr,[6,12,18]))
```

```
# 调整数组的形状
print(np.reshape(arr,[4,5]))
```

```
arr = np.array([28,30,29,31,32,30,29])
print(np.average(arr))
print(np.max(arr))
print(np.min(arr))
print(len(arr[arr>30]))
```

```
score = np.array([85,90,78,92,88])
print(np.average(score))
print(np.median(score))
print(np.std(score))
```

```
A = np.array([[1, 2],[3, 4]])
B = np.array([[5, 6],[7, 8]])
print(A + B)
print(A * B)
print(A@B)
```
