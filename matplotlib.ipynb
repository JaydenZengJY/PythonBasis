```python
# 绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['Microsoft YaHei']

# 创建一个图表，设置大小
plt.figure(figsize=(10,5))#画布

# 要绘制的数据
month = ['1月','2月','3月','4月']
sales = [100,200,120,400]
#绘制折线图
plt.plot(month,sales,
         label='产品A',
         color='red',
         linestyle='--',
         linewidth=2,
         marker='o',)
# 添加标题
plt.title('2026年销售趋势', color='blue', fontsize=20)

# 坐标轴标签
plt.xlabel('月份', fontsize=10)
plt.ylabel('销售额（万元）', fontsize=10)

#添加图例
plt.legend(loc='upper left')

#添加网格线
plt.grid(axis='y', alpha=0.5, color='blue', linestyle='--')

#设置刻度字体大小,角度
plt.xticks(rotation=30, fontsize=15)
plt.yticks(rotation=0, fontsize=15)

# 设置y轴的范围
#plt.ylim(10,160)

# 在每个数据点上显示数值
for x,y in zip(month,sales):
    #print(x,y)
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=10)

# 显示图表
plt.show()
```

```python
#饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['Microsoft YaHei']

plt.figure(figsize=(10,5))

# 要绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
colors = ['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499'] #配色

# 饼图
plt.pie(times, labels=things,
        autopct='%.1f%%',   #显示百分比
        startangle=90,   #调整初始画图角度
        colors = colors
        )

#添加标题
plt.title('一天的时间分布', color='blue', fontsize=20)

# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
#环形图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['Microsoft YaHei']

plt.figure(figsize=(10,5))

# 要绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
colors = ['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499'] #配色

# 饼图
plt.pie(times, labels=things,
        autopct='%.1f%%',   #显示百分比
        startangle=90,   #调整初始画图角度
        colors = colors, #设置配色
        wedgeprops={'width':0.5},  #圆环宽度
        pctdistance=0.7  # %数字的位置
        )

#添加标题
plt.title('一天的时间分布', color='blue', fontsize=20)
plt.text(0,0,'总计：\n100%', ha='center', va='center', fontsize=20)
# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
#爆炸式饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['Microsoft YaHei']

plt.figure(figsize=(10,5))

# 要绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
colors = ['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499'] #配色
explode = [0.1,0,0,0,0] #设置突出块的位置
# 饼图
plt.pie(times, labels=things,
        autopct='%.1f%%',   #显示百分比
        startangle=90,   #调整初始画图角度
        colors = colors,
        explode = explode,
        shadow = True,
        )

#添加标题
plt.title('一天的时间分布', color='blue', fontsize=20)

# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 散点图
import matplotlib.pyplot as plt
import random
from matplotlib import rcParams


rcParams['font.sans-serif'] = ['Microsoft YaHei']

# 创建一个图表，设置大小
plt.figure(figsize=(10,8))#画布

# 要绘制的数据
x = []
y = []
for i in range(1000):
    tmp = random.uniform(0,10)
    x.append(tmp)
    tmp2 = 2*tmp + random.gauss(0,2)
    y.append(tmp2)
# 绘制折线图
plt.scatter(x,y,
            color='green',
            alpha=0.7,
            s=20,
            label='数据')
# 添加标题
plt.title('x变量与y变量的关系', color='blue', fontsize=20)

# 坐标轴标签
plt.xlabel('x自变量', fontsize=10)
plt.ylabel('y因变量', fontsize=10)

#添加图例
plt.legend(loc='upper left')

#添加网格线
plt.grid(True, alpha=0.5, color='blue', linestyle='--')

#设置刻度字体大小,角度
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 设置y轴的范围
plt.ylim(0,30)

plt.plot([0,10],[0,20],color='red',linewidth=2,linestyle='--')

# 显示图表
plt.show()
```

```python
# 箱线图
import matplotlib.pyplot as plt
from matplotlib import rcParams

data = {
    '语文': [82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学': [75, 80, 79, 93, 88, 82, 87, 89, 92],
    '英语': [70, 72, 68, 65, 78, 80, 85, 90, 95]
}

plt.figure(figsize=(8, 6))
plt.boxplot(data.values(), tick_labels=data.keys())

plt.title("各科成绩分布（箱线图）")
plt.ylabel("分数")
plt.grid(True, axis='y', linestyle='--',alpha=0.5)
plt.show()
```
