```python
# 1.类和对象
# 1.1 类的三要素
# 1.1.1 类名：(如：人类)
# 1.1.2 属性：对象的特征描述，用来说明是什么样的（身高体重）
# 1.1.3 方法（学习，说话..）

# 2.定义类
# class 类名: #类名必须符合标识符规定，遵循大驼峰命名
#    代码块
# 举例：洗衣机类
class Washer:
    height = 800 #类属性
#查看类属性：类名.属性名
print(Washer.height)
#新增类属性：类名.属性名 = 值
Washer.width = 450
print(Washer.width)

# 3.创建对象
# 3.1格式：对象名 = 类名()
wa = Washer()
print(wa) # 显示对象的地址
# 第二次实例化
wa2 = Washer()
print(wa2) # 可实例化多个对象

# 4.实例方法和实例属性
# 4.1实例方法
# 由对象调用，至少有一个self参数，执行实例方法的时候，自动将调用该方法的对象赋值给self
class Washer1:
    height = 800
    def wash(self): # self 为当前调用方法的对象
        print('我会洗衣服')
        print(self)
wa = Washer1()
print(wa)
#对象调用类中的方法
wa.wash()
```
