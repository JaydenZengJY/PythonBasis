```python
# 导入os模块
import os
# 一、os模块
# 作用：用于和操作系统进行交互
# 1.os.name # 指示正在使用的工作平台（返回操作系统类型）
print(os.name)
# 对于windows返回nt，对于linux返回posix
```

```python
# 2.os.getenv(环境变量名称) #读取环境变量
print(os.getenv("path"))
```

```python
# 3.os.path.split() #把目录名和文件名分离，以元组的形式接收，第一个元素是目录路径，第二个元素是文件名
print(os.path.split(r"C:\Users\LENOVO\PyCharmMiscProject\function\stdlib.ipynb"))
print(type(os.path.split(r"C:\Users\LENOVO\PyCharmMiscProject\function\stdlib.ipynb")))
```

```python
# 4.os.path.dirname #显示split分割的第一个元素，即目录
# 5.os.path.basename #显示split分割的第二个元素，即文件名
print(os.path.dirname(r"C:\Users\LENOVO\PyCharmMiscProject\function\stdlib.ipynb"))
print(os.path.basename(r"C:\Users\LENOVO\PyCharmMiscProject\function\stdlib.ipynb"))
# print(os.path.basename(r"C:\Users\LENOVO\PyCharmMiscProject\function\")) #报错：如果路径以/结尾，则返回空值，如果以\结尾，报错。
```

```python
# 6.os.path.exist() # 判断路径（文件或目录）是否存在，存在的话就返回True，不存在就返回False
print(os.path.exists(r"C:\Users\LENOVO\PyCharmMiscProject\function\stdlib.ipynb"))
# 7.os.path.isfile() #判断是否存在文件
print(os.path.isfile(r"C:\Users\LENOVO\PyCharmMiscProject\function\stdlib.ipynb"))
# 8.os.path.isdir() # 判断目录是否存在
print(os.path.isdir(r"C:\Users\LENOVO\PyCharmMiscProject\function"))
# 9.os.path.abspath() # 获取当前路径下的绝对路径
print(os.path.abspath("stdlib.ipynb"))
# 10.os.path.isabs() # 判断是否为绝对路径
print(os.path.isabs(r"C:\Users\LENOVO\PyCharmMiscProject\function"))
```

```python
# 二、sys模块
# 作用：负责程序根python解释器的交互
import sys
# 1.sys.getdefaultencoding() # 获取系统默认编码格式
print(sys.getdefaultencoding())

# 2.sys.path:获取环境变量的路径，跟解释器相关
print(sys.path) # 以列表的形式返回，第一项为当前所在的工作目录
# print(sys.path[0])

```

```python
# 3.sys.platform   # 获取操作系统平台名
print(sys.platform)
```

```python
# 4.sys.version   # 获取python解释器的版本信息
print(sys.version)
```
