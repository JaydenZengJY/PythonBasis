## 函数定义格式

```python
def 函数名(形参a, 形参b):
    函数体
    ...
```

示例：

```python
def add(a, b):
    return a + b
```

---

## 函数调用格式

```python
函数名(实参1, 实参2)
```

示例：

```python
add(1, 2)
```

---

## 说明

- `形参`：定义函数时写在括号里的变量，例如 `a`、`b`。
- `实参`：调用函数时传入的具体数据，例如 `1`、`2`。

````python
# 函数参数
# 1.必备参数（位置参数）
# 含义：传递和定义参数的顺序及个数必须一致
# 格式：def func(a,b)
def funa(name, name2, name3):
    print(name)
    print(name2)
    print(name3)
funa('梅西', 'C罗', '内马尔')#顺序、个数匹配
````

````python
# 2.默认参数
# 含义：为参数提供默认值，调用函数时可不传该默认参数的值
# 注意：所有的位置参数必须出现在默认参数前，包括函数定义和调用
# 格式：def func(a=12):
def funb(b=8, a=8):
    print(b)
funb() #有默认值却没设值，那么传默认值
funb(10)#若设值则传该值
````

````python
# 3.可变参数
# 含义：传入的值的数量可以改变，可多个，可不传
# 格式：def func(*args)
def func(*args):
    print(args)# 以元组形式接收
    print(type(args))
func(1,2)
# 以元组形式接收
````

````python
# 4.关键字参数
# 格式：def func(**kwargs):
def fund(**kwargs):
    print(kwargs)
    print(type(kwargs)) # 以字典形式接收
fund(name = 'x', age = 20) #传值的时候，采用 键=值 的形式
# 作用：扩展函数的功能
````

````python
# 5.函数嵌套
# 5.1嵌套调用
# 含义：在一个函数里调用另一个函数
def study():
    print('study')
def course():
    study() #调用函数
    print('basis')
#调用
study()
course()
````

````python
# 5.2嵌套定义
# 含义：在一个函数里定义另一个函数
def study(): #外
    print('study')
    def course(): #内
        print('basis') #内函数不能调用外函数！！会死循环
    course()  #内函数要调用！
study()
````

````python
# 1.作用域
# 1.1含义：指的是变量生效的范围。分为两种：全局变量和局部变量
# 1.2全局变量
# 函数外部定义的变量，在整个文件中都能用
a = 100
def test1():
    print('这是test1中的值', a)
def test2():
    a = 120 # 局部变量
    print('这是test2中的值', a)
print('调用函数前a的值', a)
test1()
test2()
print('调用函数后a的值', a)
# 1.3局部变量
# 函数内部定义的变量，从定义位置到函数定义结束位置有效
def funa():
    num = 10  #只能在该函数内部使用
    print('num:', num)
funa()
def funb():
    num = 10  #
    print('num:', num)
funb()

# 在函数内部修改全局变量的值，可以使用global关键字
# 作用：将变量声明为全局变量
# 格式：global 变量名
a = 100
def test1():
    print('这是test1中的值', a)
def test2():
    global a #修改全局变量的值
    a = 120 # 局部变量
    print('这是test2中的值', a)
print('调用函数前a的值', a)
test1()
test2()
print('调用函数后a的值', a)
# 1.4nonlocal:在嵌套函数中对上一次进行修改
````

````python
# 2.匿名函数
# 2.1基本语法
# 函数名 = lambda 形参 ：返回值（表达式）
# 调用： 结果 = 函数名（实参）

# 普通函数
def add(a, b):
    return a + b
print(add(1, 2))
# 匿名函数
add = lambda a,b:a+b #a,b为匿名函数的形参，a+b是返回值的表达式
# lambda不需要写return来返回值
print(add(1, 3))

#2.2 lambda的参数形式
# 函数名 = lambda 形参 ：返回值（表达式）
# 2.2.1无参数
func = lambda : 'haha'
print(func())
# 2.2.2 一个参数
fund = lambda name: name
print(fund('bb'))
# 2.2.3 默认参数
fune = lambda name,age = 18:(name,age)
#有默认值的参数在无默认值参数后面！！！
print(fune('HAHA'))
print(fune('HAHA', 21))
# 2.2.4 关键字参数
funf = lambda **kwargs: kwargs
print(funf(name = 'haha', age = 18))
````

````python
# 2.3 lambda结合if判断
a = 5
b =  8
print('a比b小') if a<b else print('a大于等于b') #三目运算

comp = lambda a,b : 'a比b小' if a<b else print('a大于等于b')
# lambda只能实现简单逻辑

# 3.内置函数
# 3.1查看所有的内置函数
import builtins
#print(dir(builtins))
# 大写字母开头：一般为内置常量名；小写字母开头：一般为内置函数名
# 3.2内置函数一
# 3.2.1 abs():返回绝对值
print(abs(-1))
# 3.2.2 sum():求和
#print(sum(123)) #报错：整型不是可迭代对象
print(sum({1.1,2.9,3})) #元组，列表，集合

# 3.3 内置函数二
# 3.3.1 min()最小值
# 3.3.2 max()最大值
print(min(1.1,2.9,3))
print(max(1.1,2.9,3))
print(min(-8,5,key=abs)) #先求绝对值，再比较
# 3.3.3 zip() 将可迭代对象作为参数，将对象中对应的元素打包成一个个元组
li = [1,2,3]
li2 = ['a','b','c']
print(zip(li,li2)) #需取出对象
# 一：通过for循环取出
for i in zip(li,li2):
    print(i) #若元素不一致，以短的为准
# 二：转换成列表打印
print(list(zip(li,li2))) #必须是可迭代对象
````

````python
# 3.3.4 map()：映射
#map(func, iter1) # func:自己定义的函数。iter:要放进去的可迭代对象
l2 = [1,2,3]
def func(x):
    return x*5
mp = map(func, l2) #只写函数名，无()
print(list(mp))

# 3.3.5 reduce()
from functools import reduce
#reduce(function, sequence)# function函数必须有两个参数
# sequence为序列（可迭代）
li3 = [1,2,3,4]
def add(x,y):
    return x+y
res = reduce(add, li3)
print(res)
````
