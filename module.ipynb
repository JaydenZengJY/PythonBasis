```python
# 1.模块
# 1.1含义：一个py文件就是一个模块，即导入一个模块本质上是执行一个py文件
# 1.2分类：
# 1.2.1内置模块：
# 如：random、time、os、logging、直接导入即可使用
# 1.2.2第三方库:
# 在cmd： pip install + module
# 1.2.3自定义模块：自己定义的
# 3.3导入方式
# 3.3.1 import + 模块名
# 语法：
# 导入模块：import 模块名
# 调用功能：模块名.功能名
import demo
print(demo.name)
demo.func()

# 3.3.2 from 模块名 import 功能1，功能2
from demo import funa
funa() #直接输入功能

# 3.3.3 from ... import *
# from 模块名 import *
from demo import * #全导入
funa()
func()
```

```python
# 3.3.4
# 1.as给模块起别名
# 语法: import 模块名 as 别名
import demo as de
# 调用
de.func()
# 打印模块中的name变量
print(de.name)
```

```python
# 2. as 给功能起别名
# 语法： from 模块 import 功能 as 别名
from demo import func as f,name
f()
print(name)
```

```python
# 3.4 内置全局变量__name__
# 3.4.1 语法：
#if __name__ == "__main__":
# 3.4.2 作用
# 用来控制py文件在不同的应用场景执行不同的逻辑
# 3.4.3 __name__
# 1.文件在当前文件执行（自己执行自己）：__name__ == "__main__"
# 2.文件被当作模块被其他文件导入：__name__ == 模块名
#  被当作模块导入时，__name__ == "__main__"下面的代码不会被显示出来
import demo1
demo1.test()
```
