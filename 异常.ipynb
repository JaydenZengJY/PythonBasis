```python
# 异常：一个事件，这个事件在程序执行过程中发生，影响了程序的正常执行
# print(a) # NameError: name 'a' is not defined
# print('a') # a
# a = 10
# print(a) # 10

# 异常处理的目的： 让程序在有异常时，仍能正常运行
# 异常处理格式一： try....except
# 法1：
try:
    print(a)
except: # 基类异常
    print('出现错误')
b = 10
print(b)

# 法2：
try:
    print(a)
except NameError as e: # 指定捕获异常类型为NameError，取别名为e
    print(e)
c = 10
print(c)

# 法3：
try:
    print(a)
except Exception as e: # 万能异常Exception 可捕获任意异常
    print(e)

# 法4：多分支异常
try:
    print(a)
except IndexError as e:
    print(e)
except KeyError as e:
    print(e)
except NameError as e:
    print(e)
```

```python
# 异常捕获格式二：try except else
# else 只有在没有异常时才会执行的代码，可把except理解为if，except和else不同时执行
dic = {'name':'zs'}
try:
    print(dic['name'])
except Exception:
    print('出现错误')
else:
    print('没有捕获到异常')
```

```python
# 异常捕获格式三：try （except） finally
try:
    print(a)
except NameError:
    print('出现错误')
finally:
    print('haha') #有无错误都执行
```

```python
# (完整)异常捕获格式四：try（必选） except（二选一） else(可选) finally（二选一）
try:
    n = int(input('请输入一个整数：'))
    print(10/n)
except ValueError:
    print('请输入正确的数据')
except Exception as e: # 万能捕获异常
    print('未知错误 %s'%e)
else:
    print('没有异常时才会执行的代码')
finally:
    print('无论有无异常，都会执行的代码') #有无错误都执行
```
